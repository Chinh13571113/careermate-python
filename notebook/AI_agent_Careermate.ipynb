```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfv = TfidfVectorizer(min_df=3, max_features=None,
                      strip_accents='unicode', analyzer='word', token_pattern=r'\w{1,}',
                      ngram_range=(1, 3),
                      stop_words='english')

#Filling Nans with empty string
data_jd_clean['Skills'] = data_jd_clean['Skills'].fillna('')
```

```python
# Fitting the TF-IDF on the "Skill" text
tfv_matrix = tfv.fit_transform(data_jd_clean["Skills"])
```

```python
tfv_matrix
```

```python
tfv_matrix.shape
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

# Compute the sigmoid kernel
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
```

```python
sig[0]
```

```python
# Reverse mapping of indices and skills
indice = pd.Series(data_jd_clean.index, index=data_jd_clean['Skills']).drop_duplicates()
```

```python
indice
```

```python
print(indice["JavaScript; React; Node.js; Express.js; MongoDB; Docker; Kubernetes; AWS; GraphQL; REST APIs"])
```

```python
def give_rec(skill,sig=sig):
  # Find the index corresponding to the skill by checking if the skill string is contained in the index
  # Using .iloc[0] to get the first match if multiple skills contain the substring
  try:
      idx = indice[indice.index.str.contains(skill, case=False, na=False)].iloc[0]
  except IndexError:
      return "Skill not found in the index." # Return a message if no skill is found

  # Get the pairwise similarity score
  sig_score = list(enumerate(sig[idx]))
  # Sort the skills
  sig_score = sorted(sig_score, key=lambda x: x[1], reverse=True)
  # Score of 10 most similar skill
  sig_score = sig_score[1:11]
  #skill indice
  skill_indice = [i[0] for i in sig_score]
  # Return the top 10 most similar skill
  return data_jd_clean['Skills'].iloc[skill_indice]
```

```python
give_rec("JAVA-SPRING BOOT, NODEJS, PYTHON")
```

```python
data_jd_clean["combined_text"] = (
    data_jd_clean["Title"].fillna("") + " "
    + data_jd_clean["Skills"].fillna("") + " "
    + data_jd_clean["Responsibilities"].fillna("")
)
data_jd_clean.head(3)
```

```python
def get_gemini_embedding(text: str) -> np.ndarray:
    """Extract vector embedding from model Gemini."""
    text = text.strip().replace("\n", " ")
    if not text:
        return np.zeros(768)

    response = genai.embed_content(
        model="models/embedding-001",
        content=text,
        task_type="semantic_similarity",
    )
    return np.array(response["embedding"])
```

```python
embeddings = []
for text in tqdm(data_jd_clean["combined_text"], desc="Generating embeddings"):
    emb = get_gemini_embedding(text)
    embeddings.append(emb)

embeddings = np.vstack(embeddings)
print("Embeddings shape:", embeddings.shape)
```

```python
dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(embeddings)
faiss.write_index(index, "jd_embeddings.faiss")

# Lưu dataframe gốc để map kết quả
data_jd_clean.to_csv("jd_data.csv", index=False)
```

```python
import pandas as pd
from sentence_transformers import SentenceTransformer
```

```python
data_jp = pd.read_csv("../agent_core/data/job_postings.csv",encoding='latin-1')
data_jp.head()
```

```python
data_jp.isnull().sum()
```

```python
#rename column id to JobID
data_jp.rename(columns={'id':'JobID'}, inplace=True)
data_jp.head()
```

```python
#drop other columns not needed
data_jp_clean = data_jp[["JobID","title","skills","address"]]
data_jp_clean.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
#step 1: Generate embeddings for job postings
model = SentenceTransformer('all-MiniLM-L6-v2')
job_postings_embeddings = model.encode(data_jp_clean["Content"].tolist(), show_progress_bar=True)
```

```python
#step 2: Use cosine similarity to find similar job postings
content_similarity = cosine_similarity(job_postings_embeddings)
```

```python
# Function to get top N similar job postings
def get_similar_job_postings(job_id, top_n=5):
    # Find the index of the job posting
    idx = data_jp_clean.index[data_jp_clean['JobID'] == job_id].tolist()[0]
    # Get similarity scores for this job posting
    sim_scores = list(enumerate(content_similarity[idx]))
    # Sort the job postings based on similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    # Get the indices of the top N similar job postings
    sim_indices = [i[0] for i in sim_scores[1:top_n+1]]
    # Return the top N similar job postings
    return data_jp_clean.iloc[sim_indices]
```
